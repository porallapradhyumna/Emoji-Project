```python
from bs4 import BeautifulSoup as bs
import requests

base_url = "https://unicode.org/emoji/charts/full-emoji-modifiers.html"
all_emoji_codes = []
```

```python
def get_Data():
    base_url = "https://unicode.org/emoji/charts/full-emoji-modifiers.html"
    all_emoji_codes = []
    response = requests.get(base_url)
    soup = bs(response.content,'html.parser')
    rev_table = soup.findAll("td",attrs={"class","code"})
    for td in rev_table:
        code = td.find("a").text
        code = code.replace("+","000")
        code =code.split(" ")
        all_emoji_codes.append(code)
    return all_emoji_codes

all_emoji_codes = get_Data()
```

```python
def get_emojis_unicode(url):
    base_url = url
    response = requests.get(base_url)
    soup = bs(response.content,'html.parser')
    rev_table = soup.findAll("td",attrs={"class","chars"})
    Skin_emojis_ls = []
    Skin_unicode_ls = []
    Skin_hex_ls = []
    for i in range(len(rev_table)):
        txt = rev_table[i].getText()
        #print(txt.encode('unicode_escape'))
        Skin_unicode_ls.append(txt.encode('unicode_escape'))
        Skin_emojis_ls.append(txt)
        Skin_hex_ls.append(txt.encode())
    return(Skin_emojis_ls,Skin_unicode_ls,Skin_hex_ls)
```

```python
Skin_emojis = get_emojis_unicode("https://unicode.org/emoji/charts/full-emoji-modifiers.html")
import pandas as pd
Emojis = {
    'Emoji' : Skin_emojis[0],
    'Unicode' : Skin_emojis[1],
    "Hex" : Skin_emojis[2],
}
df = pd.DataFrame(Emojis)
print(df)
df.to_csv('Skin_Toned_Emojis.csv')
```

```python
def Get_Make_File_Name(url,parent):
    import os
    parent = parent
    if not (os.path.exists(parent)):
        os.mkdir(parent)
    base_url = url
    response = requests.get(base_url)
    soup = bs(response.content,'html.parser')
    rev_table = soup.findAll("th",attrs={'class','cchars'})
    list_filename = []
    for i in rev_table:
        name = i.getText()
        if  name not in list_filename and name != "Browser":
            list_filename.append(name)
            path =os.path.join(parent,name)
            if not (os.path.exists(path)):
                os.mkdir(path)
    return list_filename
```

```python
list_filename = Get_Make_File_Name("https://unicode.org/emoji/charts/full-emoji-modifiers.html","SkinData")
```

```python
def Download_Emojis(url):
    import base64
    import os
    response = requests.get(url)
    soup = bs(response.content,'html.parser')
    body = soup.find_all("tr")
    ls =[x[0] for x in os.walk("SkinData")]
    for i in body[3:-1]:
        #td =body[3].select("img")
        file_name = i.select("a")[0].getText()
        #print(file_name)
        imgs = i.select("img")
        directories = ls[1:len(imgs)+1]
        for id in range(len(imgs)):
            head,data = imgs[id]["src"].split(",")
            file_ext = head.split(';')[0].split('/')[1]
            filename = os.path.join(directories[id],file_name + '.' + file_ext)
            data = base64.b64decode(data)
            with open(filename, 'wb') as f:
                f.write(data)
Download_Emojis(base_url)
```
